```python
#1)Plot a Line Chart: Plot y = 3x + 2 for x in the range 0 to 10. Set the x-axis and y-axis labels, and give the plot a title. 


import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = 3 * x + 2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = 3x + 2')
plt.title('Line Chart: y = 3x + 2')
plt.grid(True)
plt.show() 
```

```python
#2)Scatter Plot with Customization: Create a scatter plot of 50 random points. Change the color of the points, increase their size, and add a grid to the plot. 

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', s=100)
plt.grid(True)
plt.show()
```

```python
#3)Bar Chart with Annotations: Draw a vertical bar chart for five categories of your choice. Annotate the value on top of each bar. 


categories = ('A', 'B', 'C', 'D', 'E')
values = (10, 20 , 30, 40, 50)
plt.figure()
bars = plt.bar(categories, values, color='skyblue')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             str(bar.get_height()), ha='center', va='bottom')
plt.title("Bar Chart with Annotations")
plt.show()
```

```python
#4)Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same axes, each with different colors and line styles. Add a legend.


x = np.linspace(0, 2*np.pi, 100)
plt.Figure()
plt.plot(x, np.sin(x), 'r-', label='sin')    # red solid
plt.plot(x, np.cos(x), 'g--', label='cos')   # green dashed
plt.plot(x, np.exp(x/3), 'b:', label='exp')  # blue dotted
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
#5)Histogram with Custom Bins: Generate 1000 random numbers from a normal distribution and plot their histogram with 30 bins. Change the bar color and add axis labels.

data = np.random.rand(1000)
plt.Figure()
plt.hist(data, bins=30, color = "Orange", edgecolor = "Black")
plt.xlabel("Values")
plt.ylabel("Frequences")
plt.title("Histogram with bins")
plt.show()
```

```python
##6)Subplots Grid: Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and histogram, each in a different subplot. Add an overall (supertitle) for the figure. 

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 20]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
data = np.random.randn(100)

# Create subplots grid
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
fig.suptitle("2x2 Subplots Grid")

# Line plot
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(categories, values)
axs[0, 1].set_title("Bar Chart")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title("Scatter Plot")

# Histogram
axs[1, 1].hist(data, bins=20, color='green')
axs[1, 1].set_title("Histogram")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# 7)Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using made-up data. 


months = np.arange(1,13)
product1 = np.random.randint(3, 15, size = 12).cumsum()
product2 = np.random.randint(3, 12, size = 12).cumsum()
product3 = np.random.randint(4, 10, size = 12).cumsum()
plt.figure()
plt.stackplot(months, product1, product2, product3,
              labels = ("Product 1, Product 2, Product s3"))
plt.legend(loc ="upper left")
plt.title("Stack Plot")
plt.show
```

```python
#8)Boxplot Using DataFrame: Given a DataFrame of student exam scores for three subjects, visualize the data distribution with a boxplot.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = {
    'Math': np.random.randint(50, 100, 20),
    'Science': np.random.randint(45, 95, 20),
    'English': np.random.randint(40, 90, 20)
}

df = pd.DataFrame(data)

df.boxplot()
plt.title("Student Exam Scores Distribution")
plt.ylabel("Scores")
plt.show()
```

```python
#9)Pie Chart with Explode: Create a pie chart showing the market share of four companies. Explode the largest segment and display percentage values on the chart.

import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [25, 40, 20, 15]

largest_index = market_share.index(max(market_share))
explode = [0.1 if i == largest_index else 0 for i in range(len(companies))]

plt.pie(market_share, labels=companies, autopct='%1.1f%%', explode=explode, shadow=True, startangle=90)
plt.title("Market Share of Companies")
plt.show()
```

```python
#10)Axis and Legend Customization: For a random line plot, reverse the x-axis, move the legend outside the plot, and increase legend font size.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.random.rand(100)

plt.plot(x, y, label="Random Line", color='blue')

plt.gca().invert_xaxis()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Axis and Legend Customization")

plt.tight_layout()
plt.show()

```
